```python
import numpy as np
import matplotlib.pyplot as plt

def generate_synthetic_raman_spectrum(raman_shifts, peak_intensities, broadening_factor):
    spectrum = np.zeros(len(raman_shifts))
    
    for shift, intensity in zip(raman_shifts, peak_intensities):
        spectrum += intensity * np.exp(-0.5 * ((raman_shifts - shift) / broadening_factor) ** 2)
    
    return spectrum

# Parameters for the synthetic spectrum
raman_shifts = np.linspace(0, 3000, 1000)  # Raman shifts in cm^-1
peak_intensities = [100, 200, 150]  # Intensity of each peak
broadening_factor = 20  # Adjust this to control the broadening

# Generate the synthetic Raman spectrum
synthetic_spectrum = generate_synthetic_raman_spectrum(raman_shifts, peak_intensities, broadening_factor)

# Plot the synthetic spectrum
plt.plot(raman_shifts, synthetic_spectrum)
plt.xlabel('Raman Shift (cm^-1)')
plt.ylabel('Intensity')
plt.title('Synthetic Raman Spectrum')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import gaussian

def generate_synthetic_spectrum(shifts, intensities, broadening_factor=10, shift_offset=0):
    spectrum = np.zeros(len(shifts))
    for shift, intensity in zip(shifts, intensities):
        shift += shift_offset  # Apply a shift offset if needed
        broadening = broadening_factor * (shift / 1000)  # Adjust broadening based on shift
        gaussian_peak = gaussian(len(shifts), broadening)
        shifted_peak = np.roll(gaussian_peak, int(shift))
        spectrum += intensity * shifted_peak
    return spectrum

# Parameters for the synthetic spectrum
shifts = np.linspace(0, 3000, 1000)  # Shifts in cm^-1
intensities = [100, 200, 150]  # Intensity of each peak

# Generate the synthetic spectrum
synthetic_spectrum = generate_synthetic_spectrum(shifts, intensities, broadening_factor=20, shift_offset=100)

# Plot the synthetic spectrum
plt.plot(shifts, synthetic_spectrum)
plt.xlabel('Shift (cm^-1)')
plt.ylabel('Intensity')
plt.title('Synthetic Spectrum')
plt.show()
```
